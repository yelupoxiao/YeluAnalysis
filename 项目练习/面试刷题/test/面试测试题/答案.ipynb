问题:

1. 统计每个城市每年土地成交的平均的单位价格（unit_price）

2. 统计每个城市每年土地成交的平均加权价格 （sum(price)/sum(area）)

3. 在1的结果上补全每个城市缺失年份的数据，向前填充

4. 将1中的平均单位价格标准化

5. 按type统计unit_price的数值分布情况

```python
import numpy as np
import pandas as pd

file_path = "./land_data.xlsx"
df = pd.read_excel(file_path, encoding='gbk')
pd.set_option('display.max_columns', None)
df1 = df.copy()
df1.head().append(df.tail())
```

```python
# 1.统计每个城市每年土地成交的平均的单位价格
res1 = df1[['city_name', 'year', 'unit_price']].groupby(['city_name','year']).mean()
res1
```

```python
# 2.统计每个城市每年土地成交的平均加权价格
res2 = df1[['city_name', 'year', 'area', 'price']].groupby(
    ['city_name','year']).apply(lambda x: np.sum(x['price']) / np.sum(x['area']))
res2
```

```python
# 3. 在1的结果上补全每个城市缺失年份的数据，向前填充
res3 = res1.unstack().stack(dropna=False).fillna(method='ffill')
res3
```

```python
# 4. 将1中的平均单位价格标准化
# Z-Score标准化
res4 = res1.apply(lambda x: (x - np.mean(x)) / np.std(x))
res4
```

```python
# 5. 按type统计unit_price的数值分布情况
res5 = df1[['type', 'unit_price']].groupby('type').describe()
res5
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': np.random.randint(1, 10, 10),
    'B': np.random.randint(1, 10, 10),
    'C': np.random.randint(1, 10, 10)
})
df
```

```python
# 按B列的值升序排序
df.sort_values(by='B')
```

```python
# 先按B列值升序排序,再按A列的值升序排序
df.sort_values(by=['B','A'])
```

```python
# 先按B列的值降序排序,再按A列的值降序排序
df.sort_values(by=['B','A'],ascending=False)
```

```python
# 先按B列的值降序排序,再按A列的值升序排序
df.sort_values(by=['B','A'],ascending=[False,True])
```

```python
# 先按B列的值升序排序,再按A列的值降序排序
df.sort_values(by=['B','A'],ascending=[True,False])
```
